```python
from google.cloud import firestore
from google.oauth2 import service_account

# ðŸ” Firestore setup
cred = service_account.Credentials.from_service_account_file("serviceAccountKey2.json")
db = firestore.Client(credentials=cred)
collection_name = "questions"

def verify_firestore_data(limit=5):
    try:
        # Fetch documents from the collection
        docs = db.collection(collection_name).stream()
        doc_list = list(docs)
        total_docs = len(doc_list)

        print(f"\nâœ… Total documents in '{collection_name}': {total_docs}")

        print(f"\nðŸ“„ Showing first {limit} documents:\n" + "-"*40)
        for doc in doc_list[:limit]:
            data = doc.to_dict()
            print(f"ID: {doc.id}")
            print(f"Question: {data.get('text')}")
            print(f"Chapter: {data.get('chapter')}")
            print(f"Options: {[opt.get('text') for opt in data.get('option_list', [])]}")
            print("-" * 40)

    except Exception as e:
        print(f"âŒ Error fetching data: {e}")

# ðŸ” Run verification
verify_firestore_data()
```

## 5.Metadata-Schema creation

```python
import os
from google.cloud import firestore
from google.oauth2 import service_account
from collections import defaultdict

# ðŸ” Firestore setup
cred = service_account.Credentials.from_service_account_file("INSERTION/serviceAccountKey2.json")
db = firestore.Client(credentials=cred)
questions_collection = "questions"
schema_collection = "schema"

def get_category_info(question_id):
    # M.1.1.S.1 format
    try:
        parts = question_id.split('.')
        if len(parts) != 5 or parts[0] != 'M' or parts[3] != 'S':
            print(f"âš ï¸ Skipping ID with invalid format: {question_id}")
            return None, None, None
        
        category = int(parts[1])
        subcategory = int(parts[2])
        return category, subcategory, parts[4]
    except Exception as e:
        print(f"âš ï¸ Error processing question ID {question_id}: {str(e)}")
        return None, None, None

def create_schema():
    try:
        print("ðŸ” Starting schema creation...")
        
        # Initialize counters
        category_stats = defaultdict(lambda: {
            'total_questions': 0,
            'subcategories': defaultdict(int)
        })
        
        # Get all questions
        print("ðŸ“¥ Fetching questions from Firestore...")
        questions = db.collection(questions_collection).stream()
        question_count = 0
        processed_count = 0
        
        # Process each question
        for question in questions:
            question_count += 1
            if question_count % 100 == 0:
                print(f"Processing question {question_count}...")
                
            question_id = question.id
            category, subcategory, _ = get_category_info(question_id)
            
            if category is not None:
                processed_count += 1
                category_stats[category]['total_questions'] += 1
                category_stats[category]['subcategories'][subcategory] += 1
        
        print(f"ðŸ“Š Processed {question_count} questions, successfully categorized {processed_count}")
        
        if question_count == 0:
            print("âŒ No questions found in the collection!")
            return
        
        # Create schema document
        schema_data = {
            'categories': {},
            'total_questions': sum(stats['total_questions'] for stats in category_stats.values())
        }
        
        # Map category numbers to names
        category_names = {
            1: "Arithmetic and Number Theory",
            2: "Algebra",
            3: "Time, Work and Speed",
            4: "Geometry and Mensuration",
            5: "Logical Reasoning",
            6: "Probability and Combinatorics",
            7: "Financial Mathematics"
        }
        
        # Map subcategory numbers to names
        subcategory_names = {
            1: "Number System",
            2: "HCF and LCM",
            3: "Decimal Fractions",
            4: "Simplification",
            5: "Square and Cube Roots",
            6: "Average",
            7: "Problems on Numbers",
            8: "Problems on Ages",
            9: "Surds and Indices",
            10: "Logarithms",
            11: "Percentage",
            12: "Profit and Loss",
            13: "Ratio and Proportion",
            14: "Partnership",
            15: "Chain Rule",
            16: "Pipes and Cisterns",
            17: "Time and Work",
            18: "Time and Distance",
            19: "Boats and Streams",
            20: "Problems on Trains",
            21: "Alligation and Mixture",
            22: "Simple Interest",
            23: "Compound Interest",
            24: "Area",
            25: "Volume and Surface Area",
            26: "Races and Games",
            27: "Calendar",
            28: "Clocks",
            29: "Stocks and Shares",
            30: "Permutations and Combinations",
            31: "Probability",
            32: "True Discount",
            33: "Bankers Discount",
            34: "Heights and Distances",
            35: "Odd Man Out and Series"
        }
        
        # Format the schema data
        for category_num, stats in category_stats.items():
            category_name = category_names.get(category_num, f"Category {category_num}")
            schema_data['categories'][category_name] = {
                'total_questions': stats['total_questions'],
                'subcategories': {}
            }
            
            for subcategory_num, count in stats['subcategories'].items():
                subcategory_name = subcategory_names.get(subcategory_num, f"Subcategory {subcategory_num}")
                schema_data['categories'][category_name]['subcategories'][subcategory_name] = count
        
        print("\nðŸ“ Schema data prepared. Uploading to Firestore...")
        
        # Upload schema to Firestore
        schema_ref = db.collection(schema_collection).document('question_distribution')
        schema_ref.set(schema_data)
        
        print("âœ… Schema created successfully!")
        print(f"ðŸ“Š Total questions: {schema_data['total_questions']}")
        print("ðŸ“‹ Categories:")
        for category, data in schema_data['categories'].items():
            print(f"  - {category}: {data['total_questions']} questions")
            for subcategory, count in data['subcategories'].items():
                print(f"    - {subcategory}: {count} questions")
                
    except Exception as e:
        print(f"âŒ Error creating schema: {str(e)}")
        raise

if __name__ == "__main__":
    create_schema() 
```
